```python
import scrapy
from scrapy.crawler import CrawlerProcess

class QuotesSpider(scrapy.Spider):
    name = 'quote'
    start_urls = [
        'https://www.orami.co.id/ap/sgm?rsq=sgm',
    ]
    
    custom_settings = {
        'FEED_FORMAT': 'csv',
        'FEED_URI': 'quotes.csv'
    }

    def parse(self, response):
        for quote in response.css('div.wrap-widget-detail'):
            yield {
                'Nama': quote.css('div.prod-name a::text').get().replace("\n", ""),
                'Harga': quote.css('div.widget-price div.onsale div.normal-price::text').get(),
            }

        next_page = response.css('li.next a::attr("href")').get()
        if next_page is not None:
            yield response.follow(next_page, self.parse)
            
# run spider
process = CrawlerProcess()
process.crawl(QuotesSpider)
process.start()
```
